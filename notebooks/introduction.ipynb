StarxETP, affiliated with [StarX](http://starxfinance.com/#1) is a startup with a vested interest in offering Exchange Traded Products (ETPs) for invested to earn profits that are relatively higher than simply investing in individual assets or risk free assets (e.g. securities bonds).

To that end, the __objectives__ of the discovery phase aims to: 
1. realize the data Extraction, Transformation, and Load (ETL) process for requesting data from popular and trusted market capitalization data sharing platforms (e.g. Coindesk), 
1. perform general portfolio analysis to understand the standard concepts relavant expected returns and ETP design and management methods, and 
1. then use those findings to plan an algorithm for developing an ETP index and ETP weighted asset rebalancing strategy.

To that end, 2 key notebooks that are supported by respective libraries offer the insigts into the aforementioned aspects. 
1. [Market capitalazination and data discovery](./etpDataDiscovery.ipynb) with simple time-series plots
1. [Expected returns](./etpReturns.ipynb) calculations relevant for an optimal ETP
1. Elements that contribute to the [ETP index and portfolio optimzation](./etpPerformIndex.ipynb)

## General Instructions for using the Notebooks
The evaluation code and results are presented through Jupyter Notebook cells ([learn about it](https://www.codecademy.com/article/how-to-use-jupyter-notebooks-py3))
* Click on the link to open any or all of the three notebooks in a new tab
* Each block in the notebook is a cell that can be executed independently but in the sequence from top to botton.
   1. Click the mouse arrow inside a cell to select it
   1. Then click on the "play" button to run the code in that cell
   1. The results from that execution will be displayed below the cell
   1. If you wish to re-run it, after changing the parameters(*), simply start from the 1st instruction listed at the top 
   1. Once a cell above is executed, you don't really need to run it again unless you wish to investigate a different dataset that is generated by a previous cell
* Refer to these instruction for [interacting with the plotly charts](https://plotly.com/chart-studio-help/zoom-pan-hover-controls/) for zoom, pan, hover over, and reset

(*) instruction for changing the parameters and other cell specific instructions are presented in the instructions above each cell
