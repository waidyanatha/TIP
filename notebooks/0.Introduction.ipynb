StarxETP, affiliated with [StarX](http://starxfinance.com/#1) is a startup with a vested interest in offering Exchange Traded Products (ETPs) for investers to earn profits that are relatively higher than simply investing in individual assets or risk free assets (e.g. securities bonds).

To that end, the __objectives__ of the discovery phase aims to: 
1. realize the data Extraction, Transformation, and Load (ETL) process for requesting data from popular and trusted market capitalization data sharing platforms (e.g. Coindesk), 
1. perform general portfolio analysis to understand the standard concepts of expected returns and indexes for optimizing the performance measures required for the ETP design and management, and 
1. then use those findings to plan an algorithm for developing an a weighted multi-asset ETP with a periodic rebalancing strategy.

To that end, 4 notebooks present the discovery work with a strategy for developing an ETP. The README.md file offers a summary of the code-based comprising classes of libraries and methods that support with the inputs and outputs for the norebooks. 
1. [Market capitalazination data discovery](./1.etpDataDiscovery.ipynb) with simple time-series plots of the market-cap trends, a simple index, and returns per unit risk 
1. [Logarithmic Ratio of Returns](./2.etpLogReturns.ipynb) (ROR) methodology for computing the returns with a comparison of simple moving average, simple standard deviation, and the actual logarithmic returns that sets the basis for determining the expected returns of an ETP and momentum trading
1. Examine several [ETP index and portfolio optimzation](./3.etpPerformIndex.ipynb) methods, including the finding top-N of the ETP based on a randomized wieghted sum of Log ROR estimation. Also examines the Average Directional Index (ADX), Sharpe, and Soratino performance indeces that will be considered for rebalancing the ETP.
1. [Algorithm for rebalancing the ETP portfolio](./4.etpBalancingPortfolio.ipynb) is still ___work in prgress___ and serves as a placeholder for the next phase.

## General Instructions for using the Notebooks
The evaluation code and results are presented through Jupyter Notebook cells ([learn about it](https://www.codecademy.com/article/how-to-use-jupyter-notebooks-py3))
* Click on the link to open any or all of the three notebooks in a new tab
* Each block in the notebook is a cell that can be executed independently but in the sequence from top to botton.
   1. Click the mouse inside a cell to select that cell
   1. Then click on the "play" button in the top menu bar to run the code in that cell
   1. The results from that execution will be displayed below the cell
   1. If you wish to re-run it, after changing the parameters(*), simply start from the 1st instruction listed at the top 
   1. Once a dependent preceeding cell above is executed (e.g. one that retrieves the data required for the current cell), you don't need to run the preceeding cell again, unless you wish to investigate a different dataset that is generated by a previous cell
* Refer to these instruction for [interacting with the plotly charts](https://plotly.com/chart-studio-help/zoom-pan-hover-controls/) for zoom, pan, hover over, and reset charts

(*) instruction for changing the parameters and other cell specific instructions are presented in the instructions above each respective cell
